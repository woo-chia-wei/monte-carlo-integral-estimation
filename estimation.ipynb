```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Estimate PI

```python
sample_size = 2000
df = pd.DataFrame({
    'x': np.random.rand(sample_size),
    'y': np.random.rand(sample_size)
})
df['under_curve'] = df.apply(lambda row: np.sqrt(row['x'] ** 2 + row['y'] ** 2) <= 1, axis=1)

estimated_area = df['under_curve'].mean()
estimated_pi = estimated_area * 4

fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(df['x'], df['y'], color=df['under_curve'].map({True:'green', False:'red'}), alpha=0.5)

t = np.linspace(0, np.radians(360), num=1000)
ax.plot(np.sin(t), np.cos(t), color='white', ls='--')

ax.set_title('')
ax.set_title(f'Equation is x^2 + y^2 = 1, Estimated PI is {estimated_pi}')
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim((-1.25, 1.25))
ax.set_ylim((-1.25, 1.25))
ax.set_facecolor('black')
plt.show()
```

# Estimate e

```python
sample_size = 2000
df = pd.DataFrame({
    'x': np.random.rand(sample_size) + 1,
    'y': np.random.rand(sample_size)
})
df['under_curve'] = df.apply(lambda row: row['y'] <= 1/row['x'], axis=1)

estimated_area = df['under_curve'].mean()
estimated_e = 10 ** (np.log10(2) / estimated_area)

fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(df['x'], df['y'], color=df['under_curve'].map({True:'green', False:'red'}), alpha=0.5)

x = np.linspace(1/2, 3, num=100)
y = 1/x
ax.plot(x, y, color='white', ls='--')

ax.set_title(f'Equation is y = 1/x, Estimated e is {estimated_e}')
ax.set_aspect('equal')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim((0, 3))
ax.set_ylim((0, 2))
ax.set_facecolor('black')
plt.show()
```
